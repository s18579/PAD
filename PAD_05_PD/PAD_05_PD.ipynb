Wgraj zawartość pliku orders.csv do ramki danych.
Przyjrzyj się danym, wykorzystując poznane metody (describe, info, head) zanalizuj dane.

Następnie:

a.	Zamień order_date na format daty. Upewnij się, że w rezultacie data będzie w formacie: rok-miesiąc-dzień;

b.	Sprawdź ile i jakie unikalne wartości znajdują się w kolumnie T-shirt category;

c.	Wyczyść kolumnę tshirt_category normalizując nazwy produktów. Upewnij się, że dla każdej kategorii funkcjonuje jedna nazwa, np. zamiast white i wh jest tylko jedna wersja white. Skorzystaj z klasy string i metody apply;

d.	Rozdziel kolumnę tshirt category na 3 kolumny: tshirt_type, która będzie mówiła o tym czy jest to koszulka, bluza czy coś innego, tshirt_size, która będzie informowała o rozmiarze koszulki, i tshirt_colour, która będzie informowała o kolorze. Skorzystaj z metody apply.

e.	Znajdź zamówienia złożone między 31 grudnia 2014 a 2 sierpnia 2016.


```python
import pandas as pd
orders = pd.read_csv('orders.csv')
orders
```

```python
# Przyjrzyj się danym, wykorzystując poznane metody (describe, info, head) zanalizuj dane.
print('Describe: '+str(orders.describe))
print('Info: '+str(orders.info))
print('Head: '+str(orders.head))
# =================================
```

```python
# a) Zamień order_date na format daty. Upewnij się, że w rezultacie data będzie w formacie: rok-miesiąc-dzień;
orders['order_date']=pd.to_datetime(orders['order_date'], format='%Y/%m/%d')
orders
# ====================
```

```python
# b) Sprawdź ile i jakie unikalne wartości znajdują się w kolumnie T-shirt category;
orders.groupby('tshirt_category')['tshirt_category'].count()
# ====================
```

```python
# c) Wyczyść kolumnę tshirt_category normalizując nazwy produktów. Upewnij się, że dla każdej kategorii funkcjonuje jedna nazwa, np. zamiast white i wh jest tylko jedna wersja white. Skorzystaj z klasy string i metody apply;
orders['tshirt_category']=orders['tshirt_category'].astype(str)
orders['tshirt_category']=orders['tshirt_category'].apply(lambda x: x.replace('Wh ','White '))
orders['tshirt_category']=orders['tshirt_category'].apply(lambda x: x.replace('Bl ','Black '))
orders['tshirt_category']=orders['tshirt_category'].apply(lambda x: x.replace('Tshirt ','T-Shirt '))
orders.groupby('tshirt_category')['tshirt_category'].count()
# ====================
```

```python
# d) Rozdziel kolumnę tshirt category na 3 kolumny: tshirt_type, która będzie mówiła o tym czy jest to koszulka, bluza czy coś innego, tshirt_size, która będzie informowała o rozmiarze koszulki, i tshirt_colour, która będzie informowała o kolorze. Skorzystaj z metody apply.
def get_type(cat: str):
    str_split = cat.split()
    if len(str_split) == 1: return str_split[0]
    elif len(str_split) == 2: return cat
    else: return str_split[1]

def get_colour(cat: str):
    str_split = cat.split()
    if len(str_split) == 3: return str_split[0]
    else: return None

def get_size(cat: str):
    str_split = cat.split()
    if len(str_split) == 3: return str_split[2]
    else: return None

orders['tshirt_type']=orders['tshirt_category'].apply(lambda x: get_type(x))
orders['tshirt_size']=orders['tshirt_category'].apply(lambda x: get_size(x))
orders['tshirt_colour']=orders['tshirt_category'].apply(lambda x: get_colour(x))
orders
# ====================
```

```python
# e) Znajdź zamówienia złożone między 31 grudnia 2014 a 2 sierpnia 2016.
orders.loc[(orders['order_date'] >= '2014-12-31') & (orders['order_date'] <= '2016-08-02')]
# ====================
```

Wgraj plik customers.csv do ramki danych. Następnie:

a.	Wypisz kolumny;

b.	Porównaj rozmiar ramki z rozmiarem ramki w zadaniu 1;

c.	Zmień nazwę kolumny customerID  na customer_id upewniając się, że dane zmiana zostanie zapisana w pamięci dzięki jednemu z argumentów metody do zmiany nazwy kolumny;

d.	Połącz ramkę orders z ramką customers po kolumnie customer_id. Z której metody skorzystasz? Zostaw odpowiedź w komentarzu do zadania w kodzie (np. blok komentarzem ‘’’ ‘’’).

```python
customers = pd.read_csv('customers.csv')
customers
```

```python
# a) Wypisz kolumny;
customers.columns
# ====================
```

```python
# b) Porównaj rozmiar ramki z rozmiarem ramki w zadaniu 1;
print('\tOrders\t\tCustomers')
print('Size\t'+str(orders.size)+'\t\t'+str(customers.size))
print('Shape\t'+str(orders.shape)+'\t'+str(customers.shape))
print('Memory\t'+str(orders.memory_usage(index=True).sum())+' B\t'+str(customers.memory_usage(index=True).sum())+' B')
# ====================
```

```python
# c) Zmień nazwę kolumny customerID  na customer_id upewniając się, że dane zmiana zostanie zapisana w pamięci dzięki jednemu z argumentów metody do zmiany nazwy kolumny;
customers.rename(columns = {'customerID':'customer_id'}, inplace = True)
customers
# ====================
```

```python
# d) Połącz ramkę orders z ramką customers po kolumnie customer_id. Z której metody skorzystasz? Zostaw odpowiedź w komentarzu do zadania w kodzie (np. blok komentarzem ‘’’ ‘’’).
customers_orders = orders.merge(customers,on = 'customer_id')
customers_orders
# Uzylem metody 'merge', ktora polaczyla poprzez klucz wspolny 'customer_id" inner joinem obydwa tabelki
# ====================
```
